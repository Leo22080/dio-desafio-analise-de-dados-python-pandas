# Python para análise da dados com Pandas

```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#Carregando a base de dados
base_titanic = pd.read_csv('../datasets/titanic.csv')
```

```python
base_titanic.head(5)
```

```python
#modificando o nome das colunas
base_titanic = base_titanic.rename(columns={
    'Survived':'Sobreviveu',
    'Pclass':'Classe',
    'Name':'Nome',
    'Sex':'Sexo',
    'Age':'Idade',
    'Siblings/Spouses Aboard':'Irmaos/conjuges a bordo',
    'Parents/Children Aboard':'Pais/filhos a bordo',
    'Fare':'Tarifa'
})
```

```python
#Exibindo as colunas modificadas
base_titanic.columns
```

```python
#Total de linhas e colunas
base_titanic.shape
```

```python
base_titanic.dtypes
```

```python
#exibindo as dez últimos linhas
base_titanic.tail(10)
```

```python
base_titanic.describe()
```

```python
base_titanic['Classe'].unique()
```

```python
#Passageiros idosos(idade >= 60)
idosos = base_titanic.loc[base_titanic['Idade'] >= 60]
idosos
```

```python
#Média da tarifa por classe
base_titanic.groupby('Classe')['Tarifa'].mean().reset_index()
```

```python
#histograma dos passageiros classicados por idade
import matplotlib.pyplot as plt
plt.hist(x = base_titanic['Idade']);
```

```python
base_titanic['Classe'].value_counts().plot.pie(title='Divisão de classes');
```

```python
plt.scatter(x=base_titanic["Tarifa"], y = base_titanic["Idade"], color='darkmagenta');
```
