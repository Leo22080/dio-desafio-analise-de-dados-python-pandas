```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#Carregando a base de dados e tirando espaços em brando do Dataframe
data_census = pd.read_csv('../datasets/census.csv', skipinitialspace = True)
```

```python
#Limpando 583 registros sem o atributo "native-country"
data_census.drop(data_census[data_census["native-country"] == '?'].index, inplace=True)
```

```python
data_census["native-country"].value_counts()
```

```python
#Média de idade por raça por pais de origem
data_census.groupby([data_census['native-country'], 'race'])['age'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
#Raça x tempo de estudo
data_census.groupby('race')['education-num'].mean().plot.barh()
plt.xlabel("education-num")
plt.ylabel("race");
```

```python
data_census['marital-status'].value_counts().plot.pie();
```

```python
data_census['race'].value_counts().plot.bar(title = "Race");
```

```python
plt.scatter(x=data_census["age"], y = data_census["hour-per-week"]);
plt.xlabel("age")
plt.ylabel("hour-per-week");
```
